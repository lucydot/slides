```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

```

```python
title_data = pd.read_csv("analyze-1.txt",delimiter=("\t"))
```

```python
title_data.columns = [c.replace(' ', '_') for c in title_data.columns]
```

```python
title_data = title_data[(title_data.Publication_Years != 2021)]
```

```python
title_data = title_data[(title_data.Publication_Years != 2022)]
```

```python
plt.scatter(title_data["Publication_Years"],title_data["Record_Count"])
plt.ylabel("Paper count")
plt.xlabel("Publication year")
plt.title("Papers with \"Computational\" in their title (Web of Science)")
```
